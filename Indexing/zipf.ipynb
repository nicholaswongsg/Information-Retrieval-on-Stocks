```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Set your CSV file path
csv_path = "/Users/nicholaswong/Desktop/Github/Information-Retrieval-on-Stocks/Classification/NER_with_ensemble_sentiment.csv"

# Load the dataset
df = pd.read_csv(csv_path)

# Tokenize and normalize text in 'ner_text_cleaned'
all_words = []
for text in df['ner_text_cleaned'].dropna():
    words = re.findall(r'\b\w+\b', text.lower())  # Lowercase and extract words
    all_words.extend(words)

# Get word frequencies
word_freq = Counter(all_words)

# Get top 30 most frequent words
top_n = 50
most_common_words = word_freq.most_common(top_n)
words = [word for word, freq in most_common_words]
frequencies = [freq for word, freq in most_common_words]

# Plot bar chart
plt.figure(figsize=(14, 6))
plt.bar(words, frequencies)
plt.title(f"Top {top_n} Most Frequent Words in ner_text_cleaned")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
